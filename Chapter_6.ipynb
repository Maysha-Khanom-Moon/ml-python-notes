## Understand Your Data With Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('dataset/diabetes.csv')
data.head(5)
```

### Univariate Plots

#### Histograms

```python
# Histograms
# data.hist()
data.hist(figsize=(12, 10), bins=20, edgecolor='black')
plt.show()
```

#### Density Plots

```python
# Density Plots
data.plot(kind='density', subplots=True, layout=(3, 3), figsize=(12, 10), sharex=False)
plt.show()
```

#### Box and Whisker Plots

```python
data.plot(kind='box', subplots=True, layout=(3, 3), figsize=(12, 10), sharex=False, sharey=False)
plt.show()
```

### Multivariate Plots

#### Correlation Matrix Plot

```python
names = data.columns

fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111)
cax = ax.matshow(data.corr(), vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0, 9, 1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names, rotation=90)
ax.set_yticklabels(names)
plt.show()
```

#### Scatter Plot Matrix

```python
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt

scatter_matrix(data, figsize=(12, 10))
plt.show()
```
